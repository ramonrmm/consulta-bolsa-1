## Consulta de datos de bolsa de Apple


#### Utilizamos las librerías de Python: yfinance, pandas y json

La idea es cargar los datos de una empresa, en este caso Apple, y con los datos devueltos  creamos una gráfica simple con la cotización de las acciones a lo largo del tiempo desde su salida a bolsa, además de los dividendos repartidos.

```python
# Importamos al proyecto las tres librerías que nos hacen falta
import yfinance as yf
import pandas as pd
import json
```

```python
# Hacemos una consulta de la empresa Apple
apple = yf.Ticker('AAPL')
```

### Recordatorio  
Si queremos ver el identificador de cualquier empresa, nos vamos a la web Yahoo Finances, buscamos la empresa que sea por su nombre y nos aparecerá con su identificador entre paréntesis

```python
# Descargamos el fichero apple.json para sacar datos de la empresa
# En este caso este fichero de prácticas está en IBM Cloud Object Storage
# con esta instrucción:
# !wget https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0220EN-SkillsNetwork/data/apple.json
# O también con esta:

!curl -O https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0220EN-SkillsNetwork/data/apple.json
```

```python
# Abrimos el fichero que hemos descargado y cargamos su contenido
# (que es un json) dentro de la variable apple_data como un diccionario
with open('apple.json','r') as file:
    apple_data = json.load(file)
```

```python
# Mostramos todos los datos de la empresa
apple_data
```

```python
# Por ejemplo, si queremos ver al sector al que pertenece:
apple_data['sector']
```

```python
# Almacenamos en la variable 'cotizacion_acciones_apple' la evolución de la cotización
# desde que salió a bolsa la empresa en 1980 hasta hoy, y nos devuelve un dataframes de pandas

cotizacion_acciones_apple = apple.history(period='max')
cotizacion_acciones_apple
```

```python
# Ahora vamos a añadir una columna de índices normal, 
# ya que la que tiene utiliza los índices de Date
# y entonces Date se queda como otra columna

cotizacion_acciones_apple.reset_index(inplace=True)
cotizacion_acciones_apple
```

```python
# Mostramos con plot una gráfica simple con la cotización.
# Recordamos que plot es una función que pertenece a la librería
# Matploblib y que pandas utiliza en backend

cotizacion_acciones_apple.plot(x='Date', y='Open')
```

```python
# Queremos ahora extraer los dividendos desde que Apple se inició en bolsa
# Para eso utilizamos el atributo dividends, que nos devuelve un dataframe
apple.dividends
```

```python
# Reindexamos los índices y renombramos las columnas

dividendos_apple = apple.dividends.reset_index()
dividendos_apple = dividendos_apple.rename(columns = {'Date':'Fecha','Dividends':'Dividendo'})
dividendos_apple
```

```python
# Mostramos con Plot la evolución de las cotizaciones
dividendos_apple.plot(y='Dividendo', x='Fecha')
```
